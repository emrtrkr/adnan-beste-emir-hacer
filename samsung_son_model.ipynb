```python
import os

# Klasörlerinizi kontrol edin
dataset_path = '/content/drive/My Drive/datason/'
print(os.listdir(dataset_path))
```

```python
from ultralytics import YOLO
```

```python
yaml_path = '/content/drive/My Drive/datason/dataa.yaml'
with open(yaml_path, 'r') as f:
    print(f.read())
```

```python
# YOLOv8'i temel alarak modeli başlat
model = YOLO('yolov8n.pt')  # 'n' versiyonu hızlı ve hafif model içindir

# Modeli eğit
model.train(
    data=yaml_path,        # YAML dosyasının yolu
    epochs=50,             # Eğitim yapılacak epoch sayısı
    imgsz=640,             # Görüntü boyutu
    batch=16,              # Batch size
    name='fish_detection'  # Modelin adı
)
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Modeli yükle
model = YOLO('/content/runs/detect/fish_detection2/weights/best.pt')

# Test görüntüsünü belirle
image_path = "/content/test.jpg"

# Görüntüyü oku ve modelle tahmin yap
results = model(image_path)

# Tahmin edilen görüntüyü göster
res_plotted = results[0].plot()  # Çizim yapılmış sonucu al
plt.imshow(cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
